```python
###############################################
## Import packages and specify some settings ##
###############################################
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# This makes plots show up and look nice
%matplotlib inline
sns.set(context='paper',style='white',font_scale=1.5,rc={"lines.linewidth":2.5})
sns.set_palette('muted')

# Keep this set to false unless you want to save figures and data
# (in which case you'll need to change some directories below)
SaveFigures=False
###############################################
###############################################

# Discretized time
T=5;
dt=.001; 
time=np.arange(0,T,dt)


plt.subplots(figsize=(10, 7.25))



### 1: A stable node
A=np.array([[-2,-1],[-1,-2]])
lam,_=np.linalg.eig(A)
u=np.zeros((2,len(time)))
u[:,0]=np.array([1.75,.25])
for i in range(len(time)-1):
    u[:,i+1]=u[:,i]+dt*(A@u[:,i])
    
plt.subplot(2,3,1)
plt.plot(u[0,:],u[1,:],label='solution')
plt.plot(u[0,0],u[1,0],'r.',MarkerSize=20,label='initial condition')
plt.plot(0,0,'k.',MarkerSize=20,label='fixed point')
plt.axis((-2,2,-2,2))
plt.xlabel(r'$u_1$')
plt.ylabel(r'$u_2$')
plt.title(r'$\lambda=%.0f, %.0f$'%(lam[0],lam[1])+'\nstable node')
plt.yticks([-2,0,2])
plt.xticks([-2,0,2])
#plt.legend(loc='upper left')
plt.tight_layout()


### 2: An unstable node
A=np.array([[2,1],[1,2]])
lam,_=np.linalg.eig(A)
u=np.zeros((2,len(time)))
u[:,0]=np.array([-.25,.4])
for i in range(len(time)-1):
    u[:,i+1]=u[:,i]+dt*(A@u[:,i])
    
plt.subplot(2,3,2)
plt.plot(u[0,:],u[1,:])
plt.plot(u[0,0],u[1,0],'r.',MarkerSize=20)
plt.plot(0,0,'k.',MarkerSize=20)
plt.axis((-2,2,-2,2))
plt.xlabel(r'$u_1$')
plt.ylabel(r'$u_2$')
plt.title(r'$\lambda=%.0f, %.0f$'%(lam[0],lam[1])+'\nunstable node')
plt.yticks([-2,0,2])
plt.xticks([-2,0,2])
plt.tight_layout()


### A saddle
A=np.array([[-1,-2],[-2,-1]])
lam,_=np.linalg.eig(A)
u=np.zeros((2,len(time)))
u[:,0]=np.array([1,1.2])
for i in range(len(time)-1):
    u[:,i+1]=u[:,i]+dt*(A@u[:,i])
    
plt.subplot(2,3,3)
plt.plot(u[0,:],u[1,:])
plt.plot(u[0,0],u[1,0],'r.',MarkerSize=20)
plt.plot(0,0,'k.',MarkerSize=20)
plt.axis((-2,2,-2,2))
plt.xlabel(r'$u_1$')
plt.ylabel(r'$u_2$')
plt.title(r'$\lambda=%.0f, %.0f$'%(lam[0],lam[1])+'\nsaddle')
plt.yticks([-2,0,2])
plt.xticks([-2,0,2])
plt.tight_layout()




### A stable spiral
A=np.array([[-1,10],[-10,-1]])
lam,_=np.linalg.eig(A)
u=np.zeros((2,len(time)))
u[:,0]=np.array([1.5,1.25])
for i in range(len(time)-1):
    u[:,i+1]=u[:,i]+dt*(A@u[:,i])
    
plt.subplot(2,3,5)
plt.plot(u[0,:],u[1,:])
plt.plot(u[0,0],u[1,0],'r.',MarkerSize=20)
plt.plot(0,0,'k.',MarkerSize=20)
plt.axis((-2,2,-2,2))
plt.xlabel(r'$u_1$')
plt.ylabel(r'$u_2$')
plt.title(r'$\lambda=%.0f \pm %.0fi$'%(lam[0].real,lam[0].imag)+'\nstable spiral')
plt.yticks([-2,0,2])
plt.xticks([-2,0,2])
plt.tight_layout()


### An unstable spiral
A=np.array([[1,-10],[10,1]])
lam,_=np.linalg.eig(A)
u=np.zeros((2,len(time)))
u[:,0]=np.array([.2,.5])
for i in range(len(time)-1):
    u[:,i+1]=u[:,i]+dt*(A@u[:,i])
    
plt.subplot(2,3,6)
plt.plot(u[0,:],u[1,:],label='solution')
plt.plot(u[0,0],u[1,0],'r.',MarkerSize=20,label='initial condition')
plt.plot(0,0,'k.',MarkerSize=20,label='fixed point')
plt.axis((-2,2,-2,2))
plt.xlabel(r'$u_1$')
plt.ylabel(r'$u_2$')
plt.title(r'$\lambda=%.0f \pm %.0fi$'%(lam[0].real,lam[0].imag)+'\nunstable spiral')
plt.yticks([-2,0,2])
plt.xticks([-2,0,2])
plt.tight_layout()
plt.legend(loc='center left', bbox_to_anchor=(-2.65,.65))


# Save figure if needed
if SaveFigures:
    plt.savefig('../Figures/LinearSystems1.pdf')
```
