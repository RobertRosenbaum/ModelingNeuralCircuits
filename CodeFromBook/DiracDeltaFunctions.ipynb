```python
###############################################
## Import packages and specify some settings ##
###############################################
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# This makes plots show up and look nice
%matplotlib inline
sns.set(context='paper',style='white',font_scale=1.5,rc={"lines.linewidth":2.5})
sns.set_palette('muted')

###############################################
###############################################

T=100
dt=.1
time=np.arange(0,T,dt)

# Parameters
t1=20
tau=10

# Make I with a delta function at
# time t1
I=np.zeros_like(time)
I[int(t1/dt)]=1/dt

# Forward Euler method to compute x
x=np.zeros_like(time)
x[0]=0
for i in range(len(time)-1):
    x[i+1]=x[i]+dt*(-x[i]+I[i])/tau


# Make figure
plt.subplots(figsize=(5, 4))

plt.subplot(2,1,1)
plt.plot(time,I)
#plt.xlabel('time (ms)')
plt.ylabel('I(t)')
plt.title('A',loc='left')
sns.despine()


plt.subplot(2,1,2)
plt.plot(time,x,color=sns.color_palette()[1])
plt.xlabel('t')
plt.ylabel('x(t)')
sns.despine()
plt.title('B',loc='left')
plt.tight_layout()

```

```python


T=100
dt=.1
time=np.arange(0,T,dt)

# Parameters
t1=20
tau=10

# Make I with a delta function at
# time t1
I=np.zeros_like(time)
I[int(t1/dt)]=1/dt


# Define a convolution kernel
s=np.arange(-5*tau,5*tau,dt)
k=(1/tau)*np.exp(-s/tau)*(s>=0)
k=k/(dt*np.sum(k))

# Define the solution as a convolution
x=np.convolve(I,k,'same')*dt


# Make figure
plt.subplots(figsize=(5.5, 4))

plt.subplot(2,1,1)
plt.plot(time,I)
#plt.xlabel('time (ms)')
plt.ylabel('I(t)')
plt.title('A',loc='left')
sns.despine()


plt.subplot(2,1,2)
plt.plot(time,x,color=sns.color_palette()[1])
plt.xlabel('t')
plt.ylabel('x(t)')
sns.despine()
plt.title('B',loc='left')
plt.tight_layout()


```
