```python
###############################################
## Import packages and specify some settings ##
###############################################
# Import packages
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# This makes plots show up and look nice
%matplotlib inline
sns.set(context='paper',style='white',font_scale=1.5,rc={"lines.linewidth":2.5})
sns.set_palette('muted')

###############################################
###############################################


# values of x at which to plot
xplot=np.arange(-2.2,2.2,.1)

# Define a function
def f(x):
    return x**2-1

plt.subplots(figsize=(9,3))

plt.subplot(1,2,1)
plt.plot(xplot,f(xplot),label='f(x)')
plt.axhline(y=0,color='black',label='phase line')
plt.plot(-1,0,'k.',markersize=15,label='fixed point')
plt.plot(1,0,'k.',markersize=15)
plt.arrow(-2,0,.2,0,head_width=0.5, head_length=0.2,color='black')
plt.arrow(.1,0,-.1,0,head_width=0.5, head_length=0.2,color='black')
plt.arrow(1,0,.5,0,head_width=0.5, head_length=0.2,color='black')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('A',loc='left')
plt.legend()
sns.despine()


T=5
dt=.01
x0=.6
time=np.arange(0,T,dt)
x=np.zeros_like(time)
x[0]=x0
# Forward Euler
for i in range(len(time)-1):
    x[i+1]=x[i]+f(x[i])*dt

plt.subplot(1,2,2)
plt.axvline(x=0,color='black')
plt.plot(time,x,'r',label='x(t)')
plt.plot(0,-1,'k.',markersize=15)
plt.plot(0,1,'k.',markersize=15)
plt.arrow(.0025,-2,.0025,.2,head_width=0.25, head_length=0.25,color='black')
plt.arrow(.0025,.1,.0025,-.1,head_width=0.25, head_length=0.25,color='black')
plt.arrow(.0025,1,.0025,.5,head_width=0.25, head_length=0.25,color='black')
plt.axhline(y=-1,color='black',linestyle='dashed')
plt.axhline(y=1,color='black',linestyle='dashed')
plt.ylabel('x')
plt.xlabel('t')
plt.yticks([-1,0,1])
plt.title('B',loc='left')
plt.legend()
sns.despine()
plt.tight_layout()

```
